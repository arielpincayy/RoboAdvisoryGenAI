```python
import pandas as pd

lista_tarjetas = [
        "Joy", "Oro", "Clasica", "Platinum", "Descubre", "Explora", 
        "Conquista", "Line Up", "La Comer", "Costo", "Home Depot", 
        "Affinity", "Teleton"
    ]

def asignar_tarjetas(row):
    
    vector_asignacion = {k: 0 for k in lista_tarjetas}
    
    age = row['Customer_Age']
    client_cat = row['Card_Category']
    
    if age < 18:
        return pd.Series(vector_asignacion)

    if client_cat in ['Blue', 'Silver', 'Gold', 'Platinum']:
        for card in ["Joy", "Oro", "Clasica", "Line Up", "La Comer", "Home Depot", "Affinity", "Teleton", "Costo"]:
            vector_asignacion[card] = 1
            
    if client_cat in ['Blue', 'Silver', 'Gold', 'Platinum']:
         vector_asignacion["Descubre"] = 1

    if client_cat in ['Gold', 'Platinum']:
        vector_asignacion["Explora"] = 1
        vector_asignacion["Platinum"] = 1

    if client_cat == 'Platinum':
        vector_asignacion["Conquista"] = 1
        
    return pd.Series(vector_asignacion)
# Aplicar la función a todo el dataset
df[lista_tarjetas] = df.apply(asignar_tarjetas, axis=1)

df[lista_tarjetas]
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
import torch.optim as optim
targets = lista_tarjetas 
X = df.drop(columns=targets, errors='ignore') 

X = pd.get_dummies(X, drop_first=True) 

y = df[targets].values 

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de y_train: {y_train.shape}")

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

class NeuralNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super(NeuralNetwork, self).__init__()
        self.layer1 = nn.Linear(input_size, 8)
        self.dropout = nn.Dropout(0.3)
        self.layer2 = nn.Linear(8, output_size)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = self.dropout(x)
        x = torch.sigmoid(self.layer2(x))
        return x

input_size = X_train.shape[1]
output_size = len(targets)
model = NeuralNetwork(input_size, output_size)

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

n_epochs = 50
batch_size = 32

for epoch in range(n_epochs):
    permutation = torch.randperm(X_train_tensor.size()[0])

    for i in range(0, X_train_tensor.size()[0], batch_size):
        indices = permutation[i:i+batch_size]
        batch_X, batch_y = X_train_tensor[indices], y_train_tensor[indices]

        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()

    with torch.no_grad():
        model.eval()
        test_outputs = model(X_test_tensor)
        test_loss = criterion(test_outputs, y_test_tensor)
        test_accuracy = ((test_outputs > 0.5) == y_test_tensor).float().mean()
        model.train()

    print(f"Epoch {epoch+1}/{n_epochs}, Loss: {loss.item()}, Test Loss: {test_loss.item()}, Test Accuracy: {test_accuracy.item()}")
```

```python

import torch
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

model.eval()
with torch.no_grad():
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.float32)
    outputs = model(X_test_tensor)
    loss_fn = torch.nn.BCELoss()
    loss = loss_fn(outputs, y_test_tensor).item()
    predictions = (outputs > 0.5).int()
    correct_preds = (predictions == y_test_tensor.int()).sum().item()
    total_samples = y_test_tensor.numel()
    acc = correct_preds / total_samples
print(f"Accuracy: {acc:.3f}")

for i, t in enumerate(targets):
    print(f"\n--- Target {i+1}/{len(targets)}: {t} ---")
    y_true_i = y_test_tensor[:, i].numpy()
    y_pred_i = predictions[:, i].numpy()
    cm = confusion_matrix(y_true_i, y_pred_i)
    print("Confusion matrix:\n", cm)
    unique_labels = np.unique(y_true_i)
    name_map = {0: 'No', 1: 'Sí'}
    labels = unique_labels.tolist()
    target_names = [name_map.get(l, str(l)) for l in labels]
    report = classification_report(y_true_i, y_pred_i, labels=labels, target_names=target_names, zero_division=0)
    print(report)
```
